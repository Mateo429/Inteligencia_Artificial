```python
#Importamos las librerías necesarias para analizar la base de datos.
import pandas as pd
import matplotlib.pyplot as plt
iris = "/content/iris.csv"
```

```python
class ITER: #Creamos la clase iterable
    def __init__(self, valor): #El método constructor crea el dataframe y crea variables para iterar sobre la clase.
        self.data_frame = pd.read_csv(iris)
        self.limite = self.data_frame.shape[0]
        self.dato = self.data_frame.iloc[valor]
        self.valor = valor

    def __iter__(self): #Se crea el método iterador
        return self

    def __next__(self):
        if self.valor < self.limite-2:
            resultado = self.data_frame.iloc[self.valor+1],self.data_frame.iloc[self.valor+2]
            self.valor += 1
            return resultado
        else:
            print("Error")
            raise StopIteration
```

```python
# Crear una instancia
plant = ITER(0)

# Usar un bucle
for flor in plant:
    print(flor)
```
